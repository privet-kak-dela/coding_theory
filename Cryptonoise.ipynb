```python
vals = [
    {1:11, 2:6,  4:8, 8:0},     
    {1:11, 2:4,  4:8, 8:15},    
    {1:6,  2:0,  4:13, 8:8},    
    {1:4,  2:10, 4:3, 8:7}      
]
```

если m чётное: 8m = 0,            если m нечётное: 8m = 8 

если m чётное, 4m может быть только 0 или 8 
значит m mod 4 = 0 или 2  и последние биты тогда 00 или 10

если m нечётное, 4m может быть только 4 или 12 
значит m mod 4 = 1 или 3, последние биты m 01 или 11


если mod 4 = 0, то m mod 8 = 0 или 4, последние 3 бита 000  или 100

если mod 4 = 2, то m mod 8 = 2 или 6, последние 3 бита 010  или 110 

если mod 4 = 1, то m mod 8 = 1 или 5, последние 3 бита 001  или 101

если mod 4 = 3, то m mod 8 = 3 или 7 ,последние 3 бита 011  или 111

знаем m mod 8

остается подставить 1 или 0 в начало и проверить, какое близко к m

```python
m = []
for i in range(4):
    s8 = vals[i][8]
    if close(s8, 0):
        mod2 = 0   
    else:
        mod2 = 1  
    
     
    s4 = vals[i][4]
    if mod2 == 0:         
        if close(s4, 0):
            mod4 = 0
        else:
            mod4 = 2
    else:                    
        if close(s4, 4):
            mod4 = 1
        else:
            mod4 = 3
    
    
    
    s2 = vals[i][2]
    
    if mod4 == 0:  pos = [0,4]
    if mod4 == 1:  pos = [1,5]
    if mod4 == 2:  pos = [2,6]
    if mod4 == 3:  pos = [3,7]
    
    mod8 = 0
    for x in pos:
        if close((2*x) % 16, s2):
            mod8 = x
            break
    
    
    s1 = vals[i][1]
    
    candidates = [mod8, Z16(mod8 + 8)]
    
    for cand in candidates:    
        if close(cand, s1): 
             m.append(cand)
            
m
```
